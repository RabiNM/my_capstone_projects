```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the CSV data into a Pandas DataFrame
csv_file_path = '/content/BUS_PROFILE_DATA.csv'  # Update with the correct file path
df = pd.read_csv(csv_file_path)

# Step 2: Filter the data by state
state_filter = 'CO'  # Specify the state you want to filter by
df_filtered = df[df['STATE_CD'] == state_filter]  # Filter by the STATE_CD column

# Step 3: Sort the filtered data by BUS_EST_NETWORK_SPEND in descending order
df_filtered = df_filtered.sort_values(by='BUS_EST_NETWORK_SPEND', ascending=False)

# Convert estimated network spend to thousands
df_filtered['BUS_EST_NETWORK_SPEND'] = df_filtered['BUS_EST_NETWORK_SPEND'] / 1000

# Step 4: Calculate the cumulative spend and cumulative percentage
df_filtered['Cumulative_Spend'] = df_filtered['BUS_EST_NETWORK_SPEND'].cumsum()
df_filtered['Cumulative_Percentage'] = 100 * df_filtered['Cumulative_Spend'] / df_filtered['BUS_EST_NETWORK_SPEND'].sum()

# Step 5: Create the Pareto chart with a larger size
fig, ax = plt.subplots(figsize=(18, 10))  # Increase figure size

# Bar plot for BUS_EST_NETWORK_SPEND
ax.bar(df_filtered['BUS_NAME'], df_filtered['BUS_EST_NETWORK_SPEND'], color='C0')

# Line plot for Cumulative Percentage
ax2 = ax.twinx()
ax2.plot(df_filtered['BUS_NAME'], df_filtered['Cumulative_Percentage'], color='C1', marker='D', ms=8, linestyle='-', linewidth=2)
ax2.axhline(y=80, color='r', linestyle='--')  # 80% line for Pareto principle
ax2.set_ylim(0, 100)

# Labeling
ax.set_xlabel('Business Name', fontsize=14)
ax.set_ylabel('Estimated Network Spend (in Thousands)', fontsize=14)
ax2.set_ylabel('Cumulative Percentage (%)', fontsize=14)
plt.title(f'Pareto Chart of Business Network Spend for State {state_filter}', fontsize=16)

# Rotate x-axis labels
plt.xticks(rotation=90, ha='right', fontsize=10)  # Adjust fontsize as needed
ax.tick_params(axis='x', labelrotation=90)

# Add grid lines
ax.grid(True, which='both', linestyle='--', linewidth=0.7)
ax2.grid(False)  # Turn off grid for the secondary y-axis

plt.tight_layout()  # Adjust layout to make room for rotated x-axis labels
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the CSV data into a Pandas DataFrame
csv_file_path = '/content/BUS_PROFILE_DATA.csv'  # Update with the correct file path
df = pd.read_csv(csv_file_path)

# Step 2: Filter the data by state
state_filter = 'CO'  # Specify the state you want to filter by
df_filtered = df[df['STATE_CD'] == state_filter]  # Filter by the STATE_CD column

# Step 3: Calculate the percentage of total spend to IT annual budget
df_filtered['SPEND_TO_BUDGET_RATIO'] = (df_filtered['BUS_EST_TOT_SPEND'] / df_filtered['EF_ANL_ITBGT']) * 100

# Step 4: Sort data by total spend in descending order (this is key for Pareto)
df_sorted = df_filtered.sort_values(by='BUS_EST_TOT_SPEND', ascending=False)

# Step 5: Calculate the cumulative spend and cumulative percentage
df_sorted['Cumulative_Spend'] = df_sorted['BUS_EST_TOT_SPEND'].cumsum()
df_sorted['Cumulative_Percentage'] = 100 * df_sorted['Cumulative_Spend'] / df_sorted['BUS_EST_TOT_SPEND'].sum()

# Step 6: Create the Pareto chart with enhanced visibility
fig, ax = plt.subplots(figsize=(18, 10))  # Increase figure size for better visibility

# Bar plot for BUS_EST_TOT_SPEND
ax.bar(df_sorted['BUS_NAME'], df_sorted['BUS_EST_TOT_SPEND'], color='C0')

# Line plot for Cumulative Percentage
ax2 = ax.twinx()
ax2.plot(df_sorted['BUS_NAME'], df_sorted['Cumulative_Percentage'], color='C1', marker='D', ms=8, linestyle='-', linewidth=2)
ax2.axhline(y=80, color='r', linestyle='--')  # 80% line for Pareto principle
ax2.set_ylim(0, 100)

# Labeling
ax.set_xlabel('Business Name', fontsize=14)
ax.set_ylabel('Estimated Total Spend', fontsize=14)
ax2.set_ylabel('Cumulative Percentage (%)', fontsize=14)
plt.title('Pareto Chart of Total Spend vs IT Annual Budget for State CO', fontsize=18)

# Rotate x-axis labels and set font size
plt.xticks(rotation=90, ha='center', fontsize=10)
ax.tick_params(axis='x', labelrotation=90)

# Add grid lines
ax.grid(True, which='both', linestyle='--', linewidth=0.7)
ax2.grid(False)  # Turn off grid for the secondary y-axis

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
